```python
# シミュレーション結果のJSONを表示してみよう
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
# JSON結果ファイルを開く
# JSON_DATA = "TvL_20240802174024.json"
JSON_DATA = "TvL_20240807131607.json"

with open(JSON_DATA) as f:
    json_obj = json.load(f)

for key, val in json_obj.items():
    print(key)
```

```python
capital_100 = json_obj["TURN_100"]["Capital"]
print( capital_100[0 : 10] ) # 先頭10件を見てみる
```

```python
for key, values in json_obj.items():
    if key.startswith('TURN_'):
        print(key)
        val = values["Capital"]
        print( "min={:.3f}, max={:.03f}, sum={:.03f}, mean={:.03f}".format( min(val), max(val), sum(val), sum(val)/len(val) ) )
        plt.hist(val, bins=36)
        plt.show()
```

```python
talents = json_obj["Talents"]
# print( talents[ 0 : 10 ] )

for key, values in json_obj.items():
    if key.startswith('TURN_'):
        print(key)
        val = values["Capital"]
        print( "min={:.3f}, max={:.03f}, sum={:.03f}, mean={:.03f}".format( min(val), max(val), sum(val), sum(val)/len(val) ) )
        plt.scatter( talents, val )
        plt.show()
```

```python
for key, values in json_obj.items():
    if key.startswith('TURN_'):
        print(key)
        lucky = values["Lucky"]
        val = values["Capital"]
        print( "min={:.3f}, max={:.03f}, sum={:.03f}, mean={:.03f}".format( min(val), max(val), sum(val), sum(val)/len(val) ) )
        plt.scatter( lucky, val )
        plt.show()
```

```python
for key, values in json_obj.items():
    if key.startswith('TURN_'):
        print(key)
        unlucky = values["Unlucky"]
        val = values["Capital"]
        print( "min={:.3f}, max={:.03f}, sum={:.03f}, mean={:.03f}".format( min(val), max(val), sum(val), sum(val)/len(val) ) )
        plt.scatter( unlucky, val )
        plt.show()
```

```python
for key, values in json_obj.items():
    if key.startswith('TURN_'):
        print(key)
        lucky = np.array(values["Lucky"])
        unlucky = np.array(values["Unlucky"])
        luck_val = lucky - unlucky
        val = values["Capital"]
        print( "min={:.3f}, max={:.03f}, sum={:.03f}, mean={:.03f}".format( min(val), max(val), sum(val), sum(val)/len(val) ) )
        plt.scatter( luck_val, val )
        plt.show()
```
